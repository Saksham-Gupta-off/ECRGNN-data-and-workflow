```python
# Importing Dataset
import pandas as pd

import matplotlib.pyplot as plt
 

 
df = pd.read_csv('../Dataset/sider.csv', low_memory=False)
 
# Initialize counts for each category
ones = []
zeros = []
nas = []
 
# Iterate over columns
for column in df.columns:
    ones.append(df[column].eq(1).sum())
    zeros.append(df[column].eq(0).sum())
    nas.append(df[column].isna().sum())
 
# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
width = 0.5  # Adjust as needed
 
bar_positions = range(len(df.columns))
 
# Plot each section
plt.bar(bar_positions, ones, width, label='Positive', color='green', edgecolor='black')
plt.bar(bar_positions, zeros, width, bottom=ones, label='Negative', color='orange', edgecolor='black')
plt.bar(bar_positions, nas, width, bottom=[i + j for i, j in zip(ones, zeros)], label='Missing', color='red', edgecolor='black')
 
# Customize plot
plt.xlabel('Columns')
plt.ylabel('Count')
plt.title('Distribution of label values in SIDER dataset')
plt.xticks(range(1, len(df.columns)))
plt.legend()
 
# Show the plot
plt.show()# 
print(bar_positions)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def multi_gaussian(x, y, means, std_devs):
    z = np.zeros_like(x)
    for mu, std_dev in zip(means, std_devs):
        z += np.exp(-((x-mu[0])**2 + (y-mu[1])**2) / (2*std_dev**2))
    return z

# Create a grid of points
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Define the means and standard deviations of the Gaussians
means = [(-2, -2), (0, 0), (2, 2)]
std_devs = [1, 0.5, 1]

z = multi_gaussian(x, y, means, std_devs)

# Plot the surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Set initial view angle
ax.view_init(elev=25, azim=80)

plt.show()
```
